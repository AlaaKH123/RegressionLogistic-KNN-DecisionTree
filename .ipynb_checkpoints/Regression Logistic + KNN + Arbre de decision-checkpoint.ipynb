```python
#Effacer les valeur aberrantes afficher a partier les Boxplot 
```

```python
q1_age = data['Age'].quantile(0.25) # Calculer le quantile d'ordre 0.25 de la variable 'age'.
q3_age = data['Age'].quantile(0.75) # Calculer le quantile d'ordre 0.75 de la variable 'age'.
IQR=q3_age-q1_age # Calculer l'interquantile de la variable 'age'
min_age=q1_age-1.5*IQR # Calculez la valeur minimale « Âge » qui n'est pas considérée comme une valeur aberrante 
max_age=q3_age+1.5*IQR # Calculez la valeur maximale « Âge » qui n'est pas considérée comme une valeur aberrante 
```

```python
data=data[data['Age']<=max_age] # Filtrez le dataframe "data" pour exclure les lignes où la valeur de la colonne "Age" est supérieure à la valeur maximale calculée
data=data[data['Age']>=min_age] #Filtrez la trame de données "data" pour exclure les lignes où la valeur de la colonne "Age" est inférieure à la valeur minimale calculée.
data.shape # afficher les nouvaux nombres de lignes et de colonnes.
```

```python
q1_credit_score = data['CreditScore'].quantile(0.25) #Calculer le quantile d'ordre 0.25 de la variable 'CreditScore'.
q3_credit_score = data['CreditScore'].quantile(0.75) #  Calculer le quantile d'ordre 0.75 de la variable 'CreditScore'.
IQR=q3_credit_score-q1_credit_score # Calculer l'interquantile de la variable 'CreditScore'
min_credit_score=q1_credit_score-1.5*IQR #Calculez la valeur minimale « CreditScore » qui n'est pas considérée comme une valeur aberrante 
max_credit_score=q3_credit_score+1.5*IQR # Calculez la valeur maximale « CreditScore » qui n'est pas considérée comme une valeur aberrante 
```

```python
data=data[data['CreditScore']<=max_credit_score] 
data=data[data['CreditScore']>=min_credit_score] 
data.shape 
```

```python
q1_NumOfProducts = data['NumOfProducts'].quantile(0.25) 
q3_NumOfProducts = data['NumOfProducts'].quantile(0.75) 
IQR=q3_NumOfProducts-q1_NumOfProducts
min_NumOfProducts=q1_NumOfProducts-1.5*IQR 
max_NumOfProducts=q3_NumOfProducts+1.5*IQR 
```

```python
data=data[data['NumOfProducts']<=max_NumOfProducts]
data=data[data['NumOfProducts']>=min_NumOfProducts] 
data.shape 
```

```python
#Division des matrices matrices des features et matrices pour la variables target 
x = data.iloc[:, 0:10]
y = data.iloc[:,10]
```

```python
# Sélectionner les colonnes à encoder en dummies
cols_to_encode = ['Geography', 'Gender']

# Appliquer la méthode one-hot encoding
x = pd.get_dummies(data=x, columns=cols_to_encode, drop_first=True)

# Afficher les 5 premières lignes de la dataset encodée
x.head()
```

```python
x.shape
```

```python
X=x.values
Y=y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler 
scaler= StandardScaler() #creates an instance of the StandardScaler() class and assigns it to the variable scaler, which can then be used to normalize the data
X_train_sc=scaler.fit_transform(X_train)
X_test_sc=scaler.transform(X_test) #utilse les données enregistrées dans fit (Min , Max)
```

```python
# standardisation des données
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.fit_transform(X_test)

x_train = pd.DataFrame(X_train_sc)
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
lr = LogisticRegression()
lr.fit(X_train_sc,y_train) 
#dans cette etape on a appliquer la regression logistic 
```

```python
y_pred_lr = lr.predict(X_test_sc) #prediction a partir de l algorithme 
```

```python
print("Confusion matrix:\n")
print(confusion_matrix(y_test,y_pred_lr)) # matrice de confusion a pour parametres les donnes predit et les donnes relles 
print('Accuracy score {0:.3f}'.format(accuracy_score(y_test,y_pred_lr))) # nbre des pred correcte / tot des prediction
print('Precision score {0:.3f}'.format(precision_score(y_test,y_pred_lr))) # TP/(TP+FP)
print('Recall score {0:.3f}'.format(recall_score(y_test,y_pred_lr))) # TP/(TP+FN)
```

```python
print(classification_report(y_test,y_pred_lr)) # la fonction classification_report calcul plusieur coeff de performence 
```
